```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from statsmodels.formula.api import ols
from sklearn import  metrics
```

Rozważmy zbiór Boston

```python
boston = datasets.load_boston()
# print description
print(boston.DESCR)
```

```python
print(boston.keys())
print(boston.data.shape)
print(boston.feature_names)
```

```python
bos = pd.DataFrame(boston.data)
bos.head()
```

```python
bos.columns = boston.feature_names
bos.head()
```

```python
bos.isnull().sum()
```

```python
print(boston.target[0:5])
bos['PRICE'] = boston.target
bos.head()
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
plt.hist(bos['PRICE'], bins=30)
plt.xlabel("House prices in $1000")
plt.show()
```

```python
sns.set(style="darkgrid")
correlation_matrix = bos.corr().round(2)

# Generate a mask for the upper triangle
mask = np.zeros_like(correlation_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(data=correlation_matrix, mask=mask, cmap=cmap, vmin=-1, vmax=1,
            square=True,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax, annot=True);
```

```python
model = ols("PRICE ~ RM", bos).fit()
# Print the summary
print((model.summary2()))
```

```python
model = ols("PRICE ~ CRIM + ZN + I(2**INDUS) + ZN:INDUS+ I(INDUS ** 2.0)", bos).fit()
# Print the summary
print((model.summary2()))
```

```python
metrics.r2_score(bos.PRICE, model.predict(bos))
```
