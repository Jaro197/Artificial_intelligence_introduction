```python
import numpy as np
from scipy import stats
from scipy.stats import multivariate_normal
import pandas as pd
import matplotlib.pyplot as plt
import math
import seaborn as sns
from sklearn import datasets
```

# Zadanie 1 
Policz korelacje: 
<ul>
<li>pearson</li>
<li>spearman</li>
<li>kendall</li>
</ul>

dla zbioru:

```python
f = lambda x: (x)
x_tr = np.linspace(0., 5, 200)
y_tr = f(x_tr)
x = stats.uniform(1, 3).rvs(100)
y = f(x) + stats.norm(0, 0.1).rvs(len(x))
```

```python
plt.figure(figsize=(6, 6));
axes = plt.gca()
axes.set_xlim([0, 5])
axes.set_ylim([0, 5])
plt.plot(x_tr, y_tr, '--k');
plt.plot(x, y, 'ok', ms=3);
plt.show()
```

```python
corr = {}
corr['pearson'], _ = stats.pearsonr(x,y)
corr['spearman'], _ = stats.spearmanr(x,y)
corr['kendall'], _ = stats.kendalltau(x,y)
print(corr)    
```

# Zadanie 6
Proszę wygenerować dane tak by 1. i 3. współrzędna były skorelowane a pozostałe nie.
Proszę zwizualizować tą zależność za pomocą 

- sns.pairplot
- sns.heatmap

```python
n = 1000
x1 = stats.norm(0, 1).rvs(n)
x2 = stats.norm(0, 1).rvs(n)
x4 = stats.norm(0, 1).rvs(n)
X = np.stack((x1, x2, 2 * x1, x4 ), 1)
```

```python
df=pd.DataFrame(X)
sns.pairplot(df)
plt.show()
```

```python
sns.set(style="darkgrid")

corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmin=-1, vmax=1,
            square=True, xticklabels=5, yticklabels=5,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.show()
corr
```
