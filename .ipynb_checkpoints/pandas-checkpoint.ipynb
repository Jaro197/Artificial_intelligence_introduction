## Reguła Trzech Sigm
Reguła Trzech Sigm dla danego rozkładu normalnego $N(mean,\sigma)$ oznacza, że w przedziale $[mean-3\sigma,mean+3\sigma]$ znajduje się 99.7\% wszystkich obserwacji.

Napisz program, który:

  * rysuje gęstość rozkładu normalnego o zadanych parametrach $mean$ i $\sigma$,
  * wylicza pole pod krzywą
  * zaznacza obszar po krzywą

Policz prawdopodobieństwa:

  * $P(X \in [mean-\sigma,mean+\sigma])$
  * $P(X \in [mean-2\sigma,mean+2\sigma])$
  * $P(X \in [mean-3\sigma,mean+3\sigma])$



```python
from scipy.stats import norm
from scipy import integrate
```

```python
def show_SD(axis, xi, yi, text):
    '''Show the area covered by 1/2/3 SDs'''
    
    fc = '#DDDDDD'
    axis.plot(x,y)
    axis.fill_between(xi, yi, facecolor=fc)
    
    axis.text(0, 0.05, text, horizontalalignment='center', fontsize=25)
    axis.set_xlim([-3.5, 3.5])
    axis.set_ylim([-0.0, 0.5])
    axis.set_yticks([])
```

```python
nd = norm()
x = np.linspace(-3.5, 3.5,100)
x1 = np.linspace(-1,1)
x2 = np.linspace(-2,2)
x3 = np.linspace(-3,3)

y = nd.pdf(x)
y1 = nd.pdf(x1)
y2 = nd.pdf(x2)
y3 = nd.pdf(x3)
fig, axs = plt.subplots(1,3, sharey=True, figsize=(10, 5))
show_SD(axs[0], x1, y1, round(norm(0, 1).cdf(1) - norm(0, 1).cdf(-1), 3))    
show_SD(axs[1], x2, y2, round(norm(0, 1).cdf(2) - norm(0, 1).cdf(-2), 3))    
show_SD(axs[2], x3, y3, round(norm(0, 1).cdf(3) - norm(0, 1).cdf(-3), 3))  
plt.show()
```

```python
def pdf(mean=0, std=1, a=-3, b=3):
    fig, ax = plt.subplots()
    x = np.linspace(-3, 3, 100)
    y = np.linspace(a, b, 100)
    ax.plot(x, norm.pdf(x, loc=mean, scale=std))
    ax.fill_between(y, norm.pdf(y, loc=mean, scale=std), alpha=0.2, color='k')
    
    pro = norm(mean, std).cdf(b) - norm(mean, std).cdf(a)
    ax.text((b+a)/2-0.5, 0.02, round(pro,3), fontsize=20)
    
    plt.show()
```

```python
pdf(0, 1, -1, 1)
```

```python
pdf(0, 1, -2, 2)
```
