```python
import numpy as np
import scipy.stats as st
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
from scipy import optimize
```

# Regresja liniowa

Naszym Celem jest empityczne sprawdzenie twierdzenia:

Niech $A \in \mathbb{R}^{N\times K}$ będzie macierzą która przekształca $\mathbb{R}^{K}$ w $\mathbb{R}^N$, i niech $y \in \mathbb{R}^{K}$.

Wtedy punkt $x_0 \in \mathbb{R}^K$ spełnia

$$
x_0 = \mathrm{argmin} \{ x \in \mathbb{R}^K   \colon \| Ax - y \|^2 \}
$$

wtedy i tylko wtedy gdy

$$
A^TA x_0 = A^Ty.
$$


Proszę wykres funkcji w kształcie 

$
f(x)=x^2
$ 

na odcinku $[0,3]$ i go narysować.

Będzie to model, który będziemy chcieli znaleźć.

```python
f = lambda x: x**2
x_tr = np.linspace(0., 3, 200)
y_tr = f(x_tr)
plt.figure(figsize=(6, 6))
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_tr, y_tr, '--k')
plt.show()
```

Proszę wylosować kilka punktów z otoczenia funkcji $f(x)=x^2$.

Na ich podstawie będziemy wykonywać naszej presykcji.

```python
x = np.array([.2, .5, .8, .9, 1.3, 1.7, 2.1, 2.7])
y = f(x) + np.random.randn(len(x))
plt.figure(figsize=(6, 6))
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_tr, y_tr, '--k')
plt.plot(x, y, 'ok', ms=10)
plt.show()
```

Nauczmy prosty model regresji liniowej

$$
y=a\cdot x + b
$$

na naszych danych.

Oraz narysujmy prostą regresji.

```python
# We create the model.
lr = lm.LinearRegression()
# We train the model on our training dataset.
lr.fit(x[:, np.newaxis], y);
print(lr.coef_)
print(lr.intercept_)
```

Definiujemy naszą funkcję regresji

```python
f_lr = lambda x: lr.coef_ * x + lr.intercept_
```

```python
x_f_lr = np.linspace(0., 3, 200)
y_f_lr = f_lr(x_tr)
plt.figure(figsize=(6, 6))
axes = plt.gca()
axes.set_xlim([0, 3])
axes.set_ylim([0, 8])
plt.plot(x_tr, y_tr, '--k')
plt.plot(x_f_lr, y_f_lr, 'g')
plt.plot(x, y, 'ok', ms=10)
plt.show()
```

Teraz możemy wykonać predykcję dla punktu 
$$
x_0 = 1.5
$$

```python
point = np.array([1.5])
y_point = lr.predict(point[:, np.newaxis])
print(y_point)
```

```python
plt.figure(figsize=(6, 6))
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_tr, y_tr, '--k')
plt.plot(x_f_lr, y_f_lr, 'g')
plt.plot(x, y, 'ok', ms=10);
plt.plot(point, y_point, 'or', ms=10)
plt.show()
```

Możemy wypisać miarę dopasowania $R^2$. Narazie nie wnikamy w to co ona oznacza.

```python
from sklearn.metrics import r2_score
print(lr.score(x[:, np.newaxis], y))
print(r2_score(lr.predict(x[:, np.newaxis]), y))
```

# Zadanie 2
Zapoznaj się z funkcją <tt>optimize.fmin_cg</tt>
https://docs.scipy.org/doc/scipy-0.19.0/reference/generated/scipy.optimize.fmin_cg.html
i znajdż minum funkcji

$$
f(x,y) = (x+1)^2+y^2
$$

```python
def f(u):
    x, y = u
    return (x + 1)**2 + y**2

x0 = np.asarray((1, 1))
result = np.array(optimize.fmin(f, x0, xtol=0.00000001, ftol=0.00000001))
print(result)
np.isclose(result, np.array([-1, 0]))
```

# Zadanie 3
Zdafiniuj funkcję opisującą błąd kwadratowy:

$$
\mathrm{compute\_error}(a,b;(x,y)) =  \sum_{i=1}^{n} ( y_i - (ax_i+b) )^2.
$$

```python
def compute_error(u):
    a, b = u
    return sum((y - (a * x + b)) ** 2)
```

# Zadanie 4
Zminimalizuj funkcją $\mathrm{compute\_error}$ po parametrach $a,b$ (numerycznie).

Czy wyniki są podobne do wyników z poprzednich zadań?

```python
x0 = np.asarray((0, 0))
np.array(optimize.fmin(compute_error, x0))
```

# Zadanie 7
Wykonaj dwie powyższe regresję na danych z wartością odstającą.


```python
f = lambda x: (x)
x_tr = np.linspace(0., 3, 200)
y_tr = f(x_tr)
x = np.array([.2, .5, .8, .9, 1.3, 1.7, 2.1, 2.7])
y = f(x) + np.random.randn(len(x)) / 5
y[1]=y[1] + 10
plt.figure(figsize=(6, 6))
axes = plt.gca()
axes.set_xlim([0, 3])
axes.set_ylim([0, 12])
plt.plot(x_tr, y_tr, '--k')
plt.plot(x, y, 'ok', ms=10)
plt.show()
```

```python
x0 = np.asarray((0, 0))
result_compute_error = np.array(optimize.fmin(compute_error, x0))
result_compute_error_1 = np.array(optimize.fmin(compute_error_1, x0))
y1 = lambda x: result_compute_error[0] * x + result_compute_error[1]
y2 = lambda x: result_compute_error_1[0] * x + result_compute_error_1[1]
```

```python
f = lambda x: x
x_tr = np.linspace(0., 3, 200)
y_tr = f(x_tr)
y_compute_error = y1(x_tr)
y_compute_error_1 = y2(x_tr)
x = np.array([.2, .5, .8, .9, 1.3, 1.7, 2.1, 2.7])
y = f(x) + np.random.randn(len(x)) / 5
y[1]=y[1]+10
plt.figure(figsize=(6, 6))
axes = plt.gca()
axes.set_xlim([0, 3])
axes.set_ylim([0, 12])
plt.plot(x_tr, y_tr, '--k')
plt.plot(x_tr, y_compute_error, label="pow")
plt.plot(x_tr, y_compute_error_1, label="abs")
plt.plot(x, y, 'ok', ms=10);
plt.legend()
plt.show()
```
