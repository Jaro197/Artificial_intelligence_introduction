```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.datasets import make_moons
from help_plot import plot_decision_regions
from sklearn.svm import SVC
```

# Polynomial Kernel

```python
X, y = make_moons(n_samples=200, noise=.1, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

# Zad.

Naucz SVM z jÄ…drem "poly" o parametrach

* degree=1, coef0=0, C=1
* degree=2, coef0=0, C=1
* degree=3, coef0=0, C=1
* degree=3, coef0=1, C=1

```python
parameters = [{"degree": 1, "coef0": 0, "C": 1}, {"degree": 2, "coef0": 0, "C": 1}, {"degree": 3, "coef0": 0, "C": 1}, {"degree": 3, "coef0": 1, "C": 1}]
```

```python
# coef0 : float, optional (default=0.0)
# Independent term in kernel function. It is only significant in â€˜polyâ€™ and â€˜sigmoidâ€™.

for param in parameters:
    poly_kernel_svm_clf = SVC(kernel="poly", degree=param["degree"], coef0=param["coef0"], C=param["C"])

    poly_kernel_svm_clf.fit(X, y)
    plot_decision_regions(X, y, poly_kernel_svm_clf)
    plt.show()
```
